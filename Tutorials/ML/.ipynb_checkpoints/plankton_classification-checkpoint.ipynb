The data consists of thousands of images (.png files) of plankton taken using a Flow Cytometer (ADD REF). Each plankton image has been classified by an expert.
```python
data_dir="/dat1/smart1n/phytoplankton classifier/" 

# list subdirectories (i.e. classes) within the year directory
def list_subdir(path):
    return [name for name in os.listdir(data_dir+path)
            if os.path.isdir(os.path.join(data_dir+path, name))]
```

As you can see, the images come in very different shapes and sizes. This is a problem for us, as our CNN classifier will require the images to have a fixed size. To get around this, let's split each image up into smaller square crops of equal size. This way, the CNN won't see the whole plankton at once, instead it will have to learn to recognise classes by seeing small segments.

Let's do this to the first Akashiwo image as an example:

```python
# extract square crops of size nxn from image:
def extract_crops(arr,n):
    rows = [i for i in range(0,arr.shape[0]-n,n)]
    # list all multiples of n in the range [0,arr.shape[1]-n]
    cols = [i for i in range(0,arr.shape[1]-n,n)]
    # list all possible combinations of rows and columns
    combos = [(r,c) for r in rows for c in cols]
    # create square images from the original image located at (r,c)
    squares = [arr[r:r+n,c:c+n] for r,c in combos]
    return squares

def extract_crops_reshape(im,n):
    
    nx=im.shape[1]//64
    ny=im.shape[0]//64
    if nx==0 or ny==0: return 
    im=im[:ny*64,:nx*64]

    im=im.reshape((ny,64,nx,64))
    im=im.swapaxes(1,2)
    im=im.reshape(-1,64,64)

    return im
    
```

Each of these smaller square images will have the class label: Akashiwo. So the CNN will have to learn to recognise textures in the smaller images that determine the class.

Note that some of the edge images are empty. This is a problem for us as there will be lots of empty images with different class labels since there are large empty borders in many of the images. Avoiding any of these making it into the dataset is hard, but we can try a crude method for reducing the number of them, we'll see that a little later.

Now that we've explored the data a little, let's go ahead and load in lots of images (extracting square crops from each) into a pandas dataframe.

```python
# load image file and split into squares of size (n,n)
def load_image(path, n):
    im = image.imread(path)
    arr = np.asarray(im)
    squares = extract_crops(arr,n)
    return squares
```

```python
# add list of squares and the class name to a dictionary with the class name as key
def add_to_dict(image, class_name):
    return {'class': class_name, 'image': image}
```

```python
# create Pandas dataframe from list of image paths and class names
def create_df(classes):
    dicts = []
    for class_name in classes:
        # list all png files in class_name subdirectory of each year directory
        pngs = []
        for year in years:
            pngs.extend([f for f in list_files(year+'/'+class_name) if f.endswith('.png')])
        for path in pngs:
            squares = load_image(path, n=64)
            for square in squares:
                dicts.append(add_to_dict(square, class_name))
    df = pd.DataFrame(dicts)
    return df
```

Now let's try getting rid of some of those empty images. If the image is empty, then its pixel intensity (varying from 0 - 1, black - white) should be fairly uniform. Therefore, we should expect these images' pixel intensities to have a smaller standard deviation than those with plankton features in. 

Let's test this by plotting a histogram from the standard deviations of each image in the dataframe:

```python
stds = []
for i in range(len(df)):
    stds.append(np.std(df.iloc[i]['image']))
plt.hist(stds,bins = 100);
```

It looks like there's a large peak of images with low standard deviation. These could be our empty images. Let's have a look at some of them to check our hypothesis...

Play around with different standard deviation thresholds, the cell will randomly plot 16 examples below this threshold. If we try a higher threshold value (e.g. 0.1) you'll see many of the images contain recognisable plankton features, we don't want to discard these! If we try a lower threshold (e.g. 0.02) then most of the images we discard look empty. This is not an exact science! Perhaps you can think of a better way to exclude these empty squares??

This example shows why you should always visualise and inspect your data before using it in a machine learning model, your model is only as good as the data you train it on!

```python
# create list of indices with standard deviations lower than threshold
threshold = 0.02
stds_array = np.array(stds)
indices = np.argwhere(stds_array<threshold)
indices = indices.flatten()
print(len(indices))
# plot 16 of these at random
fig, axs = plt.subplots(4,4, figsize = (8,8))
for i in range(4):
    for j in range(4):
        axs[i,j].imshow(df.iloc[indices[np.random.randint(0,len(indices))]]['image'], cmap = 'gray', vmin = 0, vmax = 1)
        axs[i,j].axis('off')
        axs[i,j].set_title(df.iloc[indices[np.random.randint(0,len(indices))]]['class'])
```

Ok, let's go ahead and remove these empty images from the dataframe, and tidy up the indexing

```python
# remove rows where std(image) is below a threshold
def remove_low_std(df, threshold):
    return df[df['image'].apply(lambda x: np.std(x)>threshold)]

df = remove_low_std(df, threshold=0.02)
df.reset_index(drop=True, inplace=True)
```

Let's look at the most frequently occuring classes in the dataset, the numbers show the number of square images with each class label remaining in the dataset

```python
class_rank = df['class'].value_counts()
class_rank
```

There are lots of classes, and some of them only have a very small number of images. For this tutorial, let's focus on only the 10 most frequently occuring classes

```python
# select only the top 10 classes from df and create a new dataframe with only the top 10 classes
df = df[df['class'].isin(class_rank.index[:10])]
df.reset_index(drop=True, inplace=True)
df['class'].value_counts()
```

Even with this smaller set of classes, there is a large 'class imbalance' (i.e. there are many more examples of some classes than of others). A CNN trained on this dataset is likely to perform poorly in classifying the least frequently occuring class, as it will have 'seen' fewer examples compared to the other classes. There are a number of ways to deal with class imbalance (ADD REF), here we'll use 'undersampling'. This means we'll remove samples from the frequently occuring classes until all classes have the same number of samples as the least frequently occuring, thus creating a smaller but balanced data set. 

```python
# extract N examples from each class
def extract_examples(df, N):
    examples = []
    classes = df['class'].unique()
    for class_name in classes:
        class_df = df[df['class']==class_name]
        examples.append(class_df.sample(N))
    return pd.concat(examples)

# set number of examples to be that of least common class
N = class_rank[9]

df = extract_examples(df, N)
df.reset_index(drop=True, inplace=True)
df['class'].value_counts()
```

Next, in order to train our CNN we need the class labels to be numeric, rather than those long Latin names! For this, we use 'one hot' encoding. The class of an image is represented by a vector of length 10 (the number of classes), with a 1 for the entry corresponding to that class, and filled with zeroes everywhere else. We'll create these one hot vectors later, for now let's add an integer label for each class.

```python
df['class_label'] = pd.Categorical(df['class']).codes
df
```

So, in the end our CNN will predict a vector of length 10. If it does a good job, then for an image of a Lepotcylindrus it should predict a vector with values close to zero everywhere apart from in column 6, where it should predict a value close to one.

To train our CNN we'll use 'supervised learning'. Since we have a dataset of images with the correct corresponding labels we can train our CNN by example to correctly classify the images. This is done by defining a 'loss function', which is a measure of how well the image was classified (high value = bad, low value = good) which we then seek to minimise through 'gradient descent' (ADD REF). There are different loss function for different purposes, but for classification we'll use 'categorical cross entropy'. 

To avoid overfitting or underfitting the data, we need to split our dataset into 3 partitions: training, validation, and testing:
* Training: these are the data the network will be trained on. The network's parameters will be adjusted to minimise the loss function when applied to these data.
* Validation: during training, we'll monitor performance on the validation data to see when the model stops improving. Note that the parameters aren't adjusted based on the performance on these data!
* Testing: we only use these data right at the end to test the performance of our model on data it's never seen before.

These datasets should be independent, so we'll randomly split the large dataset into these 3 smaller sets (let's use a ratio 7:2:1, training:validation:testing).

```python
# randomly select training, validation, and test sets from df with equal class distribution
def split_df(df, train_size, val_size, test_size):
    classes = df['class'].unique()
    train_df = pd.DataFrame()
    val_df = pd.DataFrame()
    test_df = pd.DataFrame()
    for class_name in classes:
        class_df = df[df['class']==class_name]
        train_df = train_df.append(class_df.sample(frac=train_size))
        val_df = val_df.append(class_df.sample(frac=val_size))
        test_df = test_df.append(class_df.sample(frac=test_size))
    return train_df, val_df, test_df

train_df, val_df, test_df = split_df(df, train_size=0.7, val_size=0.2, test_size=0.1)
train_df.reset_index(drop=True, inplace=True)
val_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```

Now that we've prepared our datasets, it's time to define our machine learning models. There is a whole zoo of different models to try, but let's try 3 simple examples here: multi-layer perceptron (MLP), a simple convolutional neural network (CNN), and a CNN with 'residual learning' in the form of some skip connections (ResNet). 

All of these models will take as input a tensor with dimensions: (image height, image width, number of image colour channels). In our case this will be (64,64,1), note if we were using colour images this would be (64,64,3) since the images would have R, G, and B channels. 

Each model will output a vector of length 10 encoding the predicted class.

The machine learning models will be defined and trained using the popular open-source machine learning APIs: Tensorflow and Keras (PyTorch is another popular API). 

```python
# define a simple CNN model to classify plankton types
def create_cnn_model(input_shape, output_shape):
    x = keras.layers.Input(shape=input_shape)
    # add a convolutional layer with a ReLU activation
    y = keras.layers.Conv2D(8, (3,3), activation='relu')(x)
    # add a max pooling layer
    y = keras.layers.MaxPooling2D(pool_size=(2,2))(y)
    # add a convolutional layer with a ReLU activation
    y = keras.layers.Conv2D(16, (3,3), activation='relu')(y)
    # add a max pooling layer
    y = keras.layers.MaxPooling2D(pool_size=(2,2))(y)
    # add a convolutional layer with a ReLU activation
    y = keras.layers.Conv2D(32, (3,3), activation='relu')(y)
    # add a max pooling layer
    y = keras.layers.MaxPooling2D(pool_size=(2,2))(y)
    # add a convolutional layer with a ReLU activation
    y = keras.layers.Conv2D(64, (3,3), activation='relu')(y)
    # add a max pooling layer
    y = keras.layers.GlobalAveragePooling2D()(y)
    # add a flatten layer

    # add a fully connected layer with a ReLU activation
    y = keras.layers.Dense(32, activation='relu')(y)
    # apply dropout to prevent overfitting
#    y = keras.layers.Dropout(0.5)(y)
    # add final output layer with a softmax activation
    y = keras.layers.Dense(output_shape, activation='softmax')(y)
    
    model = keras.models.Model(inputs=x, outputs=y)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

CNN = create_cnn_model(input_shape=(64,64,1), output_shape=10)
CNN.summary()
```

```python
# create a numpy array of validation images and labels
test_images = np.array(test_df['image'].tolist())
test_images= np.expand_dims(test_images, axis=3)
test_labels = np.array(test_df['class_label'].tolist())
# create the one-hot encoding of the validation labels
test_labels = keras.utils.to_categorical(test_labels, 10)
```

Let's train the MLP model. One 'epoch' corresponds to one iteration through the whole training dataset. The data is fed into the model in batches, after each batch the model parameters are adjusted. To monitor performance on the validation set, we'll keep track of the accuracy on the validation data. The accuracy varies from 0 to 1 (bad predictions to good predictions).

```python
# create a numpy array of training images and labels
train_images = np.array(train_df['image'].tolist())
train_labels = np.array(train_df['class_label'].tolist())
train_images= np.expand_dims(train_images, axis=3)
# create the one-hot encoding of the training labels
train_labels = keras.utils.to_categorical(train_labels, 10)


# create a numpy array of validation images and labels
val_images = np.array(val_df['image'].tolist())
val_images= np.expand_dims(val_images, axis=3)
val_labels = np.array(val_df['class_label'].tolist())
# create the one-hot encoding of the validation labels
val_labels = keras.utils.to_categorical(val_labels, 10)

# create a numpy array of validation images and labels
test_images = np.array(test_df['image'].tolist())
test_images= np.expand_dims(test_images, axis=3)
test_labels = np.array(test_df['class_label'].tolist())
# create the one-hot encoding of the validation labels
test_labels = keras.utils.to_categorical(test_labels, 10)

hist = CNN.fit(x = train_images, y = train_labels, epochs=10, batch_size=32, validation_data=(val_images, val_labels), shuffle=True)
```

```python
plt.plot(hist.history['val_accuracy'],label='Validation')
plt.plot(hist.history['accuracy'],label='Training')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy');
```

```python
print("Evaluate on test data")
results = CNN.evaluate(test_images, test_labels, batch_size=128)
print("test loss, test acc:", results)
```
